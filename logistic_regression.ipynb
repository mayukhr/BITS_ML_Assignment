```python
import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
col_names = ['sex', 'length', 'diameter', 'height', 'wholeWeight', 'shuckedWeight', 'visceraWeight', 'shellWeight', 'rings', 'legalHarvest']
df = pd.read_csv("abelone_subset.csv",header=None, names=col_names)
df.head()
```

```python
#split dataset in features and target variable
feature_cols =  ['sex', 'length', 'diameter', 'height', 'wholeWeight', 'shuckedWeight', 'visceraWeight', 'shellWeight', 'rings']
X = df[feature_cols] 
y = df.legalHarvest

# split X and y into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
y_train
```

```python
X_train
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
CM = metrics.confusion_matrix(y_test, y_pred)
# CM-> Confusion Matrix view
CM
```

```python
# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(CM), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('CM', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
## Let's check Accuracy of Logistic Regression:
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
